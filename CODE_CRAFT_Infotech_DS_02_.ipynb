# Load Packages

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Load the dataset

```python
data = pd.read_csv('age_gender.csv')
```

# Step 1: Check for duplicates

```python
duplicate_count = data.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")
```

Step 2: Summarize basic statistics of numeric columns

```python
print(data.describe())
```

# Step 3: Check unique values in categorical columns

```python
print(f"Unique values in 'ethnicity': {data['ethnicity'].unique()}")
print(f"Unique values in 'gender': {data['gender'].unique()}")
```

# Plot age distribution

```python
plt.figure(figsize=(10, 6))
plt.hist(data['age'], bins=30, color='skyblue', edgecolor='black')
plt.title('Age Distribution', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Plot gender distribution


```python
plt.figure(figsize=(8, 5))
data['gender'].value_counts().sort_index().plot(kind='bar', color=['blue', 'pink'])
plt.title('Gender Distribution', fontsize=16)
plt.xlabel('Gender (0 = Male, 1 = Female)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(ticks=[0, 1], labels=['Male', 'Female'], fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Plot ethnicity distribution

```python
plt.figure(figsize=(8, 5))
data['ethnicity'].value_counts().sort_index().plot(kind='bar', color='orange', edgecolor='black')
plt.title('Ethnicity Distribution', fontsize=16)
plt.xlabel('Ethnicity (0 to 4)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
